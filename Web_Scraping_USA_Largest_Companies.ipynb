```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/api/rest_v1/page/html/List_of_largest_companies_in_the_United_States_by_revenue"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0 Safari/537.36"
}
page = requests.get(url, headers=headers)


soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
largest_public_companies_table = soup.find('table')
```

```python
print(largest_public_companies_table)
```

```python
largest_public_companies_titles = largest_public_companies_table.find_all('th')
```

```python
largest_public_companies_titles
```

```python
largest_public_companies_table_titles = [title.text.strip() for title in largest_public_companies_titles]
```

```python
print(largest_public_companies_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = largest_public_companies_table_titles)

df
```

```python
column_data = largest_public_companies_table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv(r'C:\Users\Marvel\Documents\Python Web Scraping USA Largest Companies\USA_Largest_Public_Companies.csv', index = False)
```
